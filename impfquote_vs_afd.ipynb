```python
import pandas as pd
```

```python
# downloaded from https://www.rki.de/DE/Content/InfAZ/N/Neuartiges_Coronavirus/Daten/Impfquotenmonitoring.html
df_impfquote = (pd.read_csv("impfquote.csv")
  .rename(columns={"Zweimal":"Impfquote"})
)
df_impfquote
```

```python
# source https://dawum.de/AfD/
df_afd = (pd.read_csv("afd_percentage.csv", decimal=",")
 .rename(columns={"Unnamed: 0" : "Bundesland", "aktuelle Umfrage" : "AFD_Wähler" }) 
)
df_afd
```

```python
df = df_afd.merge(df_impfquote,on="Bundesland").set_index("Bundesland")
```

```python
df.plot.scatter(x="AFD_Wähler",y="Impfquote", title="Impfquote vs AFD Wähler nach Bundesländer")
```

```python
df[["AFD_Wähler","Impfquote"]].sort_values("Impfquote").plot.bar()
```
